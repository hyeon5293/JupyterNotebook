```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
file_name = '코로나_19발생현황__해외__20211208182203.xlsx'
df = pd.read_excel(file_name, skiprows = 1, index_col = '국가별(2)', usecols = 'B, D')
df = df.head(10)
df
```

```python
row = list(df.index)
row.reverse()
df = df.reindex(index = row)
df
```

```python
plt.figure(figsize = (13, 7))
plt.title('국가별 코로나 사망자 Top10')
plt.barh(df.index, df['사망자수 (명)'] // 10000, color = 'r', height = 0.5)
```

```python
plt.figure(figsize = (13, 8))
plt.title('국가별 코로나 사망자 Top10 (2021.12.08)')
plt.barh(df.index, df['사망자수 (명)'], color = 'r', height = 0.5)
plt.xlabel('단위 (명)')
plt.ylabel('국가')
plt.xlim(0, 900000)
for idx, rect in enumerate(df['사망자수 (명)']):
    plt.text(rect + 3000, idx, str(format(df['사망자수 (명)'][idx], ',')) + '명', va = 'center', color = 'r', size = 11)

plt.grid(axis = 'x', alpha = 0.3, ls = '--')
plt.savefig('2021_국가별 코로나 사망자 top10.png', dpi = 80)
```
